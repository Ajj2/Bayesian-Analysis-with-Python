```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
import seaborn as sns
palette = 'colorblind'
sns.set_palette(palette); sns.set_color_codes(palette)
sns.set_style('whitegrid')
```

```python
with sns.plotting_context('talk', font_scale=1.5):
    mu_params = [-1,1]
    sd_params = [0.2, 1.5]
    x = np.linspace(-5, 5, 100)
    scale=1.5
    f, ax = plt.subplots(len(mu_params), len(sd_params), sharex=True, sharey=True, figsize=(scale*6,scale*4))
    for i in range(len(mu_params)):
        for j in range(len(sd_params)):
            mu = mu_params[i]
            sd = sd_params[j]
            y = stats.norm(mu, sd).pdf(x)
            ax[i,j].plot(x, y)
            ax[i,j].plot(0, 0, 
            label="$\\mu$ = {:3.2f}\n$\\sigma$ = {:3.2f}".format(mu, sd), alpha=0)
            ax[i,j].legend()
    for x in ax[-1, :]:
        x.set_xlabel('$x$')
    for y in ax[:, 0]:
        y.set_ylabel('$pdf(x)$')
    plt.tight_layout()
    plt.savefig('normal_dist_2x2.png', dpi=300)   
```

```python
scale=1.5
with sns.plotting_context('talk', font_scale=scale):
    n_params = [1, 4]
    p_params = [0.25, 0.75]
    ns = np.arange(0, max(n_params)+1)
    f, ax = plt.subplots(len(n_params), len(p_params), sharex=True, sharey=True, figsize=(scale*6, scale*4))
    for i in range(len(n_params)):
        for j in range(len(p_params)):
            n = n_params[i]
            p = p_params[j]
            y = stats.binom(n=n, p=p).pmf(ns)
            ax[i,j].bar(ns, y, width=0.7)
            ax[i,j].set_ylim(0, 1)
            ax[i,j].plot(0, 0, label="$N = {0}$\n$\\theta = {1:3.2f}$".format(n, p), alpha=0)
            ax[i,j].legend()
    for x in ax[-1,:]:
        x.set_xlabel('$y$')
        x.set_xticks(ns)
    for y in ax[:, 0]:
        y.set_ylabel('$p(y|\\theta)$')

    plt.savefig('binomial_dist_2x2.png', dpi=300)   
```

```python
with sns.plotting_context('talk', font_scale=scale):
    params = [0.5, 3]
    theta = np.linspace(0, 1, 100)
    f, ax = plt.subplots(len(params), len(params), sharex=True, 
      sharey=True, figsize=(scale*6, scale*4))
    for i in range(len(params)):
        for j in range(len(params)):
            a = params[i]
            b = params[j]
            y = stats.beta(a, b).pdf(theta)
            ax[i,j].plot(theta, y)
            ax[i,j].plot(0, 0, label="$\\alpha$ = {:3.2f}\n$\\beta$ = {:3.2f}".format(a, b), alpha=0)
            ax[i,j].legend()
            
    for x in ax[-1,:]:
        x.set_xlabel('$\\theta$')
    for y in ax[:, 0]:
        y.set_ylabel('$p(\\theta)$')
        
    plt.savefig('beta_dist_2x2.png', dpi=300)
```

```python
sns.set()
theta_real = 0.35
trials = [0,  8, 50, 150]
data = [0, 4, 13, 48]

beta_params = [(20, 20)]
dist = stats.beta
x = np.linspace(0, 1, 100)

f, ax = plt.subplots(nrows=1, ncols=len(trials), sharex=True, sharey=True, figsize=(16,4))
ax[0].set_xlabel('$\\theta$', fontsize=14)
ax[0].set_ylabel('$probability$', fontsize=14)

for idx in range(len(trials)):
    a_prior, b_prior = beta_params[0]
    p_theta = dist.pdf(x, a_prior, b_prior)
    p_theta_given_y = dist.pdf(x, a_prior + data[idx], b_prior + trials[idx] - data[idx])
    p1 = ax[idx].plot(x, p_theta_given_y, label="$p(\\theta|y)$")
    p2 = ax[idx].plot(x, p_theta, label="$p(\\theta)$")
    ax[idx].fill_between(x, 0, p_theta_given_y, alpha=0.6)
    ax[idx].fill_between(x, 0, p_theta, alpha=0.6)

    ax[idx].axvline(theta_real, ymax=0.3, color='k')
    ax[idx].plot(0, 0, alpha=0)
    plt.xlim(0,1)
    plt.ylim(0,12) 
    ax[idx].set_title("{:d} experiments\n{:d} heads".format(trials[idx], data[idx]))
    ax[idx].legend()
    
plt.tight_layout()

# plt.savefig('B04958_01_05.png', dpi=300, figsize=(5.5, 5.5))
```

```python
def naive_hpd(post):
    sns.kdeplot(post)
    HPD = np.percentile(post, [2.5, 97.5])
    plt.plot(HPD, [0, 0], label='HPD {:.2f} {:.2f}'.format(*HPD), 
      linewidth=8, color='k')
    plt.legend(fontsize=16);
    plt.xlabel(r"$\theta$", fontsize=14)
    plt.gca().axes.get_yaxis().set_ticks([])

    
np.random.seed(1)
post = stats.beta.rvs(5, 11, size=1000)
naive_hpd(post)
plt.xlim(0, 1)
plt.savefig('B04958_01_07.png', dpi=300, figsize=(5.5, 5.5))
```

```python
np.random.seed(1)    
gauss_a = stats.norm.rvs(loc=4, scale=0.9, size=3000)
gauss_b = stats.norm.rvs(loc=-2, scale=1, size=2000)
mix_norm = np.concatenate((gauss_a, gauss_b))

naive_hpd(mix_norm)
plt.savefig('B04958_01_08.png', dpi=300, figsize=(5.5, 5.5))
```
